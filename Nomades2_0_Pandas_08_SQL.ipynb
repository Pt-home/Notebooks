```python
import pandas as pd
import sqlite3
```

```python
con = sqlite3.connect("sample_data/weather_2012.sqlite")
df = pd.read_sql("SELECT * from weather_2012 LIMIT 5", con)
df
```

```python
df = pd.read_sql("SELECT * from weather_2012", con, index_col='id')
df
```

```python
df = pd.read_sql("SELECT * from weather_2012", con, index_col=['id', 'date_time'])
df
```

```python
weather_df = pd.read_csv('sample_data/weather_2012.csv')

con = sqlite3.connect("sample_data/test_db.sqlite")

con.execute("DROP TABLE IF EXISTS weather_2012")

weather_df.to_sql("weather_2012", con)
```

```python
df = pd.read_sql("SELECT * from weather_2012", con, index_col='index')
df
```

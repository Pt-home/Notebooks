```python
f = 'sample_data/brodsky_poetry.txt'
```

```python
import re

# Load the text data
with open(f, "r", encoding='utf8') as file:
    text = file.read()
    
text = text.upper()

while text.find('  ') != -1:
  text = re.sub(r'  ', ' ', text)
while text.find('--') != -1:
  text = re.sub(r'--', '-', text)

text = re.sub(r'\s-\s', ' — ', text)
text = re.sub(r'[^-А-ЯЁ—\s]', '', text)
```

```python
#Подсчет пробелов
def count_syllables(word):
    answer = 0
    for el in "УЕЫАОЭЯИЮЁ":
        answer += word.count(el)
    return answer
```

```python
import nltk
nltk.download('punkt')

import random
```

TRIGRAMS

```python
#============= НАСТРОЙКИ ===============

#Длина рифмы
len_rhyme = 3

#Минимальное количество вариантов для одной рифмы
min_nabor = 20

#Cтрок в стихе и слогов в строке
num_lines = 16 #кратно по четыре
num_syllables = 16

#Ограничение шагов генерации набора конечных слов в строках (предотвращает зависания цикла)
max_steps_for_generations = 2000
```

```python
# Define the Markov model
premodel = nltk.probability.ConditionalFreqDist()

words = nltk.word_tokenize(text)

#Создаём словарь рифм
dict_rhyme = {}      

for i in range(len(words)-1, 1, -1):

    #Модель в обратном порядке!!!
    key = words[i-1] + " " + words[i]
    premodel[key][words[i-2]] += 1

    #Отрезаем рифму и по ней, как ключу, сохраняем пару слов, на нее заканчивающуюся
    key_rhyme = key[-len_rhyme:]

    if key_rhyme not in dict_rhyme:
        dict_rhyme[key_rhyme] = [key]
    else:
        dict_rhyme[key_rhyme].append(key)

#Удаление "коротких наборов" рифм (где мало вариантов) и мусора (пробелы, тире и дефисы в ключе)
list_keys = list(dict_rhyme.keys())
for key in list_keys:
    if len(dict_rhyme[key]) < min_nabor or " " in key or "-" in key or "—" in key:
        del dict_rhyme[key]
```

```python
print(dict_rhyme.keys())
print(len(dict_rhyme.keys()))
```

```python
model = nltk.ConditionalProbDist(premodel, nltk.MLEProbDist)
len(model)
```

```python
def generate_poem_tri(model, num_lines, num_syllables, start_condition):
     
    # Создаем список для хранения сгенерированных строк
    lines = []
   
    # Генерируем каждую строку
    for i in range(num_lines):
      
        condition = start_condition[i]
        line_words = condition.split()
        line_syllables = count_syllables(line_words[-2]) + count_syllables(line_words[-1]) 

        # Генерируем строку, пока количество слогов не достигнет заданного значения
        while line_syllables < num_syllables:

            # Получаем вероятностное распределение для предшествующего слова и генерируем его
            word = model[condition].generate()

            # Считаем количество слогов в слове
            syllable_count = count_syllables(word)
            
            # Если количество слогов превышает оставшееся количество для строки, то выбираем новое слово КОСТЫЛЬ
            if line_syllables + syllable_count > num_syllables:
                while len(word) != num_syllables - line_syllables:
                      word = random.choice(words)

            # Добавляем слово в строку и увеличиваем количество слогов
            line_words = [word] + line_words
            line_syllables += syllable_count
                        
            # Обновляем текущее условие для следующей генерации
            condition = line_words[0] + " " + line_words[1]

       # Добавляем строку в список
        lines.append(' '.join(line_words))
        
    # Возвращаем сгенерированные строки
    return lines
```

```python
def poem_create(max_steps_for_generations):

    #Cчетчик для контроля попадания в тупиковые узлы генерации
    count = 0

    #Две рифмы
    r1 = random.choice(list(dict_rhyme.keys()))
    r2 = random.choice(list(dict_rhyme.keys()))
    while r1==r2:
      r2 = random.choice(list(dict_rhyme.keys()))

    # Создаем начальное условие из случайных слов в модели
    start_condition = []
    start_condition_words = []

    for i in range(0, num_lines, 2):

      key1  = random.choice(dict_rhyme[r1])
      w1, w2 =  map(str, key1.split())

      #Если последнее слово уже было в списке последних слов - перегененрируем
      while w2 in start_condition_words:
            key1 = random.choice(dict_rhyme[r1])
            w1, w2 =  map(str, key1.split())
            
            count += 1
            if count>max_steps_for_generations:
                print('Неудачная попытка... Пробую еще раз...\n')
                return False

      start_condition.append(key1)
      start_condition_words.append(w2)

      key2  = random.choice(dict_rhyme[r2])
      w1, w2 =  map(str, key2.split())

      #Если последнее слово уже было в списке последних слов - перегененрируем
      while w2 in start_condition_words:
            key2 = random.choice(dict_rhyme[r2])
            w1, w2 =  map(str, key2.split())
            
            count += 1
            if count>max_steps_for_generations:
                print('Неудачная попытка... Пробую еще раз...\n')
                return False

      start_condition.append(key2)
      start_condition_words.append(w2)

    #Генерируем стих
    poem = generate_poem_tri(model, num_lines, num_syllables, start_condition)
    return poem
```

```python
poem = False
while not poem:
    poem = poem_create(max_steps_for_generations)

#Выводим
print("Итак, вот результат!\n")
print('\nx x x\n')
for i in range(len(poem)):
    print(poem[i])
    if (i+1)%4 == 0:
        print('\nx x x\n')
```

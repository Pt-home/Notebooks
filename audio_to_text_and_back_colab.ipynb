# 🎧 Аудіо в текст → Озвучка роботом
Цей ноутбук дозволяє:
- Завантажити аудіо (.mp3 або .wav)
- Розпізнати текст (українською)
- Зберегти файл локально або на Google Drive


```python
# 🔧 Встановлення потрібних бібліотек
!pip install numpy==1.23.5
!pip install -q git+https://github.com/openai/whisper.git
!pip install -q TTS pydub

import whisper
from pydub import AudioSegment
import os
from IPython.display import Audio, display
from google.colab import files, drive
```

```python
# 📤 Завантаження аудіофайлу
uploaded = files.upload()
audio_path = list(uploaded.keys())[0]
display(Audio(audio_path))
```

```python
# 🧠 Розпізнавання тексту (українська)
model = whisper.load_model("base")
result = model.transcribe(audio_path, language="uk")
recognized_text = result['text']
print("🔎 Розпізнаний текст:")
print(recognized_text)
```

```python
## 🗣️ Озвучка тексту — 1 варіант
from TTS.api import TTS


#model_name =  "tts_models/uk/mai/glow-tts"
model_name =  "tts_models/uk/mai/vits"

tts = TTS(model_name)

# Генерація аудіо
tts.tts_to_file(text=recognized_text,
                file_path="synthesized.wav")
display(Audio("synthesized.wav"))
```

```python
## 🗣️ Озвучка тексту — 2 варіант
!pip install gtts
from gtts import gTTS

tts = gTTS(recognized_text, lang='uk')
tts.save("gtts_output.mp3")

display(Audio("gtts_output.mp3"))
```
